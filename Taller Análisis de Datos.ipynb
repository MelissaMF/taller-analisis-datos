```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Pokemon.csv')
print("Primeras filas")
print(df.head())
```

```python
print("Información del DataFrame")
print(df.info())
```

```python
print("Estadísiticas básicas")
print(df.describe())
```

Renombrar columnas

```python
df = df.rename(columns={'#' : 'ID', 'Name': 'Nombre', 'Type 1': 'Tipo_1', 'Type 2' : 'Tipo_2', 'HP' : 'Puntos_Vida' })
df.head()
```

Filtrar pokemones tipo fuego

```python
pokemones_fuego = df[df['Tipo_1'] == 'Fire']
pokemones_fuego.head() #Imprime los 5 primeros valores
```

```python
pokemones_fuego.tail() #Imprime los 5 últimos valores
```

```python
print("Cantidad de pokemones de fuego")
print(len(pokemones_fuego)) #La función len nos permite conocer el tamaño de elementos en el dataframe
```

```python
#Mostrar sólo algunos elementos del dataframe
```

```python
print(pokemones_fuego[['Nombre', 'Tipo_2', 'Puntos_Vida']])
```

```python
#Ordenar por puntos de vida
pokemones_fuego = pokemones_fuego.sort_values('Puntos_Vida', ascending=False)
pokemones_fuego.head()
```

```python
#Ordenar por ataque de menor a mayor
pokemones_fuego = pokemones_fuego.sort_values('Attack', ascending=True)
pokemones_fuego.head()
```

Volvemos a trabajar con el dataframe inicial

```python
#Agrupar datos
estadistica_tipo = df.groupby('Tipo_1')[['Attack', 'Defense', 'Puntos_Vida']].mean()
estadistica_tipo = estadistica_tipo.sort_values('Attack', ascending=True)
estadistica_tipo
```

Graficar en Matplotlib

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=estadistica_tipo.reset_index(), x='Tipo_1', y='Attack')
plt.title('Estadísticas de Ataque por Tipo de Pokémon')
plt.xlabel('Tipo de Pokémon')
plt.ylabel('Estadística de Ataque')
plt.xticks(rotation=90)
plt.show()
```
